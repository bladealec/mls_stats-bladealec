## Simple script to join dataframes for a PowerBI visualization

```python
import pandas as pd
```

```python
all_players = pd.read_csv('../data/MLS/all_players.csv')
all_goalkeepers = pd.read_csv('../data/MLS/all_goalkeepers.csv')
```

```python
all_players.head()
```

```python
all_goalkeepers.head()
```

```python
df = pd.merge( all_players, all_goalkeepers,
                 how="left",
                 left_on= ["Club","Year"] , right_on= ["Club","Year"],
                 left_index=False, right_index=False,
                )
df.head()
```

```python
df.POS_x.unique()
```

```python
all_defenders = df[(df['POS_x'] == 'D') | (df['POS_x'] == 'D-M') | (df['POS_x'] == 'M-D')]
print(all_defenders.shape)
all_defenders.head()
```

```python
all_defenders.to_csv('../data/MLS/all_defenders.csv', index=False)
```
